```python
driver=webdriver.Chrome(r"C:\Users\DELL\Downloads\chromedriver_win32.exe")
```

```python
df=pd.DataFrame({'Title':job_title,'Location':job_location,'Company Name':company_name,'Experience':experience_required})
df
```

```python
import selenium   
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome(r"C:\Users\DELL\Downloads\chromedriver_win32.exe")
```

```python
driver.get("https://www.jagranjosh.com/")
```

```python
gk=driver.find_element(By.XPATH,'/html/body/div/div[1]/div/div[1]/div/div[6]/div/div[1]/header/div[3]/ul/li[9]/a')
gk.click()
```

```python
minister=driver.find_element(By.XPATH,'//a[@href="https://www.jagranjosh.com/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1"]')
minister.click()
```

```python
Name=[]

names=driver.find_elements(By.XPATH,'//div[@class="table-box"]/table/tbody/tr/td[2]/p')

for name in names:

    Name.append(name.text)
    
Born_died=[]

name=driver.find_elements(By.XPATH,'//div[@class="table-box"]/table/tbody/tr/td[3]/p')
for i in name:
    Born_died.append(i.text)

Term_office=[]

term=driver.find_elements(By.XPATH,'//div[@class="table-box"]/table/tbody/tr/td[4]/p[1]')
for i in term:
    Term_office.append(i.text)
    
Remarks=[]
remark=driver.find_elements(By.XPATH,'//div[@class="table-box"]/table/tbody/tr/td[5]/p')
for i in remark:
    Remarks.append(i.text)

```

```python
import pandas as pd
df=pd.DataFrame({'Minister':Name,"Date":Born_died,"Term":Term_office,"Remarks":Remarks})
df
```
